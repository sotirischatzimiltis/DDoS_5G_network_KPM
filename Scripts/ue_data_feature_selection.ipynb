```python
import pandas as pd
import numpy as np
```

```python
df_ue = pd.read_csv('../Datasets/amari_ue_data.csv')
```

```python
df_ue.head()
```

```python
features_to_remove = ['5g_tmsi', 'amf_ue_id', 'bearer_0_apn', 'bearer_0_ip',
                      'bearer_0_ipv6', 'bearer_0_pdu_session_id',
                      'bearer_0_qos_flow_id', 'bearer_0_sst', 'bearer_1_apn',
                      'bearer_1_ip', 'bearer_1_ipv6', 'bearer_1_pdu_session_id',
                      'bearer_1_qos_flow_id', 'bearer_1_sst', 'cell_X_cell_id',
                      'ran_plmn', 'registered', 'rnti', 't3512', 'tac',
                      'tac_plmn', 'ue_aggregate_max_bitrate_dl',
                      'ue_aggregate_max_bitrate_ul', 'dl_bitrate','ul_bitrate',
                      'ran_id', 'ran_ue_id', 'cell_X_ri']
df_ue = df_ue.drop(columns=features_to_remove)
df_ue['timestamp'] = pd.to_datetime(df_ue['timestamp'])
df_ue = df_ue.sort_values(by=['imeisv', 'timestamp'])
df_ue.head()
```

```python
# print how many Nan values exist in each column
nan_counts = df_ue.isnull().sum()
for column, count in nan_counts.items():
    print(f"Column '{column}': {count} NaN values")
```

```python
# Remove rows where only the first two features have values
print("Number of instances before: ",len(df_ue))
df_ue = df_ue[df_ue.iloc[:, 2:].notna().any(axis=1)].reset_index(drop=True)
print("Number of instances after: ",len(df_ue))
```

```python
# print how many Nan values exist in each column
nan_counts = df_ue.isnull().sum()
for column, count in nan_counts.items():
    print(f"Column '{column}': {count} NaN values")
```

```python
# save the rows with the nan values

# Identify rows with NaN values in any of the columns except the first one
nan_rows = df_ue[df_ue.iloc[:, 1:].isna().any(axis=1)]

# Save the rows with NaN values to a CSV file
nan_rows.to_csv('nan_rows.csv', index=False)
```

```python
# replace nan values to zero 
df_ue['cell_X_dl_mcs'] = df_ue['cell_X_dl_mcs'].fillna(0)
df_ue['bearer_1_dl_total_bytes'] = df_ue['bearer_1_dl_total_bytes'].fillna(0)
df_ue['bearer_1_ul_total_bytes'] = df_ue['bearer_1_ul_total_bytes'].fillna(0)
df_ue['cell_X_turbo_decoder_avg'] = df_ue['cell_X_turbo_decoder_avg'].fillna(0)
df_ue['cell_X_turbo_decoder_max'] = df_ue['cell_X_turbo_decoder_max'].fillna(0)
df_ue['cell_X_turbo_decoder_min'] = df_ue['cell_X_turbo_decoder_min'].fillna(0)
```

```python
# remove cell_X_ul_mcs
df_ue = df_ue.drop(columns = ['cell_X_ul_mcs'])
```

```python
# save new dataframe
df_ue.to_csv('ue_data_new.csv',index=False)
```
